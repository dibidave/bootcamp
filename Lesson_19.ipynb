```python
import pandas
import os

import bokeh.io
import bokeh.plotting

bokeh.io.output_notebook()
```

```python
CSV_FILE_PATH = os.path.join("data", "gfmt_sleep.csv")
df = pandas.read_csv(CSV_FILE_PATH, na_values="*")
```

```python
df.head()
```

```python
figure = bokeh.plotting.figure(
    frame_width=400,
    frame_height=300,
    x_axis_label="confidence when correct",
    y_axis_label="confidence when incorrect"
)
```

```python
figure.circle(
    source=df,
    x='confidence when correct',
    y='confidence when incorrect'
)

bokeh.io.show(figure)
```

```python
figure = bokeh.plotting.figure(
    frame_width=400,
    frame_height=300,
    x_axis_label="confidence when correct",
    y_axis_label="confidence when incorrect"
)
```

```python
df['insomnia'] = df['sci'] <= 16

df.groupby('insomnia')['percent correct'].median()
```

```python
x = 'confidence when correct'
y = 'confidence when incorrect'

# Add glyphs
figure.circle(
    source=df.loc[~df['insomnia'], :],
    x=x,
    y=y,
    legend_label='normal sleepers'
)

figure.circle(
    source=df.loc[df['insomnia'], :],
    x=x,
    y=y,
    color='orange',
    legend_label='insomniacs'
)

bokeh.io.show(figure)
```

```python
figure.output_backend = 'svg'

bokeh.io.save(
    figure,
    filename='insomniac_confidence_correct.svg',
    title='Bokeh plot',
    dpi=300
)
```
