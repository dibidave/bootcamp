```python
import numpy
from scipy import stats
import bokeh.io
import bokeh.plotting

bokeh.io.output_notebook()
```

```python
NUM_SAMPLES = 10
NUM_REPITITIONS = 10000
PERCENTILES = [2.5, 97.5]
NUM_SAMPLES = numpy.logspace(1, 10, 10, base=2).astype(numpy.int32)

x_values = []
y_values = []
error_high = []
error_low = []

for num_samples in NUM_SAMPLES:
    
    ratios_unsampled = numpy.empty(NUM_REPITITIONS)

    indices = list(range(num_samples))
    indices_set = set(indices)

    for i in range(NUM_REPITITIONS):

        samples = numpy.random.choice(indices, size=num_samples, replace=True)

        ratio_unsampled = len(indices_set.difference(samples))/num_samples
        ratios_unsampled[i] = ratio_unsampled
        
    x_values.append(num_samples)
    y_values.append(numpy.mean(ratios_unsampled))
    
    percentile = numpy.percentile(ratios_unsampled, PERCENTILES)
    
    error_high.append(percentile[1])
    error_low.append(percentile[0])
```

```python
figure = bokeh.plotting.figure(
    frame_height=400,
    frame_width=500,
    x_axis_label="# Samples",
    y_axis_label="Ratio Unsampled",
#     y_axis_type="log",
    x_axis_type="log"
)

figure.circle(
    x=x_values,
    y=y_values,
    size=5
)

figure.line(
    x=[min(NUM_SAMPLES), max(NUM_SAMPLES)],
    y=[1/numpy.exp(1), 1/numpy.exp(1)]
)

bokeh.io.show(figure)
```
