```python
import pandas
import os

import bokeh.io
import bokeh.plotting

bokeh.io.output_notebook()
```

```python
CSV_FILE_PATH = os.path.join("data", "collins_switch.csv")
```

```python
df = pandas.read_csv(CSV_FILE_PATH, comment="#")
df.head()
```

```python
figure = bokeh.plotting.figure(
    frame_width=400,
    frame_height=300,
    x_axis_label="IPTG (mM)",
    y_axis_label="normalized GFP expression (a.u.)",
    x_axis_type="log",
    y_axis_type="log"
)
```

```python
figure.circle(
    source=df,
    x="[IPTG] (mM)",
    y="normalized GFP expression (a.u.)"
)

bokeh.io.show(figure)
```

```python
df["error_low"] = df["normalized GFP expression (a.u.)"] - 1.96* df["sem"]
df["error_high"] = df["normalized GFP expression (a.u.)"] + 1.96* df["sem"]
```

```python
df
```

```python
figure = bokeh.plotting.figure(
    frame_width=400,
    frame_height=300,
    x_axis_label="IPTG (mM)",
    y_axis_label="normalized GFP expression (a.u.)",
    x_axis_type="log",
    y_axis_type="log"
)
```

```python
from bokeh.models import ColumnDataSource, Whisker
```

```python
source_error = ColumnDataSource(
    data=dict(
        base=df["[IPTG] (mM)"],
                lower=df["error_low"], upper=df["error_high"]))

figure = bokeh.plotting.figure(
    frame_width=400,
    frame_height=300,
    x_axis_label="IPTG (mM)",
    y_axis_label="normalized GFP expression (a.u.)",
    x_axis_type="log",
#     y_axis_type="log"
)

figure.circle(
    source=df,
    x="[IPTG] (mM)",
    y="normalized GFP expression (a.u.)"
)

figure.add_layout(
    Whisker(source=source_error, base="base",
            upper="upper", lower="lower")
)

bokeh.io.show(figure)
```
