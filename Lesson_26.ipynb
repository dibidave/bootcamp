```python
import numpy
import os
import pandas

import  iqplot

import bokeh.io
import bokeh.plotting

bokeh.io.output_notebook()
```

```python
CSV_FILE_PATH = os.path.join("data", "grant_complete.csv")
VARIABLE_1 = "beak length (mm)"
VARIABLE_2 = "beak depth (mm)"
YEARS = [1975, 2012]

df = pandas.read_csv(CSV_FILE_PATH, comment="#")
```

```python
species_mask = df["species"] == "scandens"
year_mask = df["year"].isin(YEARS)

df = df.loc[(species_mask & year_mask), ["year", VARIABLE_1, VARIABLE_2]]
```

```python
NUM_REPITITIONS = 10000

random = numpy.random.default_rng(42)

values_by_year = {
    year: {
        VARIABLE_1: df.loc[df["year"] == year, VARIABLE_1].values,
        VARIABLE_2: df.loc[df["year"] == year, VARIABLE_2].values
    }
    
    for year in YEARS
}

for i in range(NUM_REPITITIONS):
    
    pass
```

```python
figure = iqplot.ecdf(
    data=df,
    q=VARIABLE_2,
    cats="year",
    style="staircase",
    conf_int=True
)

bokeh.io.show(figure)
```
